### Objective
Analyze Building 2 electricity consumption using basic Python programming and file handling.

## Step 1: Load the Dataset
Read the CSV file manually, skip the header, and store the electricity consumption values from the first column in a list.

```
building_data=[]

with open("building_2.csv","r") as file:
    next(file)
    for line in file:
        values=line.strip().split(",")
        building_data.append(float(values[0]))

print("Total Records:",len(building_data))
```

## Step 2: Annual Electricity Consumption

The annual electricity consumption is obtained by summing all values in the dataset.

```
annual_consumption=sum(building_data)
print("Annual Electricity Consumption:",annual_consumption,"kWh")
```

## Step 3: Monthly Consumption

Group every 720 hours (30×24) and calculate monthly totals.

```
monthly_consumption=[]
hours_per_month=24*30
for i in range(0,len(building_data),hours_per_month):
    monthly_consumption.append(sum(building_data[i:i+hours_per_month]))
for i,v in enumerate(monthly_consumption,1):
    print(f"Month {i}: {v:.2f} kWh")
```

## Step 4: Daily Consumption

Group every 24 hours to calculate daily energy usage.

```
daily_consumption=[]
for i in range(0,len(building_data),24):
    daily_consumption.append(sum(building_data[i:i+24]))
for i,v in enumerate(daily_consumption[:10],1):
    print(f"Day {i}: {v:.2f} kWh")
```

## Step 5: Seasonal Consumption

Divide data into four seasonal blocks and calculate total consumption.

```
season_hours=24*30*3
winter=sum(building_data[0:season_hours])
spring=sum(building_data[season_hours:2*season_hours])
summer=sum(building_data[2*season_hours:3*season_hours])
autumn=sum(building_data[3*season_hours:4*season_hours])
print("Winter:",winter)
print("Spring:",spring)
print("Summer:",summer)
print("Autumn:",autumn)
```

## Step 6: Weekday vs Weekend

Compare weekday and weekend electricity consumption.

```
weekday_consumption=0
weekend_consumption=0
for i,val in enumerate(daily_consumption):
    if (i%7)<5:
        weekday_consumption+=val
    else:
        weekend_consumption+=val
print("Weekday:",weekday_consumption)
print("Weekend:",weekend_consumption)
```

## Step 7: Interpretation

Identify important trends from the calculated values.

```
print("Highest Month:",max(monthly_consumption))
print("Lowest Month:",min(monthly_consumption))
if weekday_consumption>weekend_consumption:
    print("Weekdays consume more electricity.")
elif weekend_consumption>weekday_consumption:
    print("Weekends consume more electricity.")
else:
    print("Equal consumption.")
```

## Step 8: Conclusion

Summarize the analysis.

```
print("Analysis completed successfully.")
print("Annual, monthly, daily, seasonal and weekday/weekend analyses were performed.")
print("The results help understand electricity consumption patterns.")
```
